## a) Identify opportunities (if any) to create a composite feature, drop a feature

```python
# creating a new feature water-cement ratio
# The water to cement ratio largely determines the strength and durability of the concrete when it is cured properly.
# The w/c ratio refers to the ratio of the weights of water and cement used in the concrete mix.
# A w/c ratio of 0.4 means that for every 100 lbs of cement used in the concrete, 40 lbs of water is added.
# Normal for ordinary concrete (sidewalks and driveways): 0.6 to 0.7
# Specified if a higher quality concrete is desired: 0.4
# A ratio of 0.3 is very stiff (unless superplasticizers are used).
# A ratio of 0.8 makes a wet and fairly weak concrete.
# The lower the w/c ratio, the stronger the concrete.
```

```python
data['wat_cem_ratio']=data['water']/data['cement']
```

```python
data.head()
```

# K Means Clustering

```python
cluster_range = range( 2, 6 )   # expect 3 to four clusters from the pair panel visual inspection hence restricting from 2 to 6
cluster_errors = []
for num_clusters in cluster_range:
  clusters = KMeans( num_clusters, n_init = 5)
  clusters.fit(data)
  labels = clusters.labels_
  centroids = clusters.cluster_centers_
  cluster_errors.append( clusters.inertia_ )
clusters_df = pd.DataFrame( { "num_clusters":cluster_range, "cluster_errors": cluster_errors } )
#clusters_df[0:15]
clusters_df
```

```python
# Elbow plot

plt.figure(figsize=(12,6))
plt.plot( clusters_df.num_clusters, clusters_df.cluster_errors, marker = "o" )
```

### elbow technique shows that there are 4 clusters in the dataset

```python
from sklearn.cluster import KMeans
cluster = KMeans( n_clusters = 4, random_state = 2354 )
cluster.fit(data)

prediction=cluster.predict(data)
data["cluster_no"] = prediction     # Creating a new column "cluster_no" which will hold the cluster id of each record
```

```python
# getting centroids
centroids = cluster.cluster_centers_
centroids
```

```python
data.boxplot(by = 'cluster_no',  layout=(2,4), figsize=(15, 10))
```

### There are many outliers on each dimension  (indicated by the black circles)
### Spread of data on each dimension (indivated by the whiskers is long ... due to the outliers)
### If the outliers are addressed, the clusters will overlap much less than right now 

```python
# Addressing outliers at group level

clean_data = data
       
def replace(group):
    median, std = group.median(), group.std()  #Get the median and the standard deviation of every group 
    outliers = (group - median).abs() > 2*std # Subtract median from every member of each group. Take absolute values > 2std
    group[outliers] = group.median()       
    return group

data_corrected = (clean_data.groupby('cluster_no').transform(replace)) 
concat_data = data_corrected.join(pd.DataFrame(data['cluster_no']))
```

```python
concat_data.boxplot(by = 'cluster_no', layout=(2,4), figsize=(15, 10))
```

# 3) Create the model

```python
data.head()
```

```python
# getting the unique values in cluster number
data['cluster_no'].unique()
```

```python
# Convert categorical variable into dummy/indicator variables. As many columns will be created as distinct values
# This is also kown as one hot coding. 
data.cluster_no=pd.get_dummies(data.cluster_no,dtype='int64')
data.head()
```

```python
data.shape
```

```python
#creating a reference of the dataframe 'data'
data1=pd.DataFrame(data)
data1.head()
```

```python
#storing the input variables in X and output variables in y
X=data.loc[:,['slag','ash','superplastic','coarseagg','fineagg','age','wat_cem_ratio','cluster_no']]
y=data.strength
```

```python
# Spliting the dataset into train and test data
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=10)
```

## simple linear regression

```python
# reating a model linear_model with LinearRegression method
linear_model=LinearRegression()
# fitting the training data
linear_model.fit(X_train,y_train)
```

```python
# storing the predicted value in y_predict
y_predict=linear_model.predict(X_test)
```

```python
# getting the r2_score
r2_score(y_test,y_predict)
```

### support vector regressor

```python
# creating an svm model with support vector regression method
svm_model=SVR(gamma='scale')
# fitting the training data
svm_model.fit(X_train,y_train)
```

```python
# storing the predicted value in y_predict
y_predict=svm_model.predict(X_test)
```

```python
# getting the r2_score
r2_score(y_test,y_predict)
```

### Decision tree regressor

```python
# creating a decision tree model with decision tree regressor method
dtree_model=DecisionTreeRegressor(random_state=8)
# fitting the training data
dtree_model.fit(X_train,y_train)
```

```python
# storing the predicted value in y_predict
y_predict=dtree_model.predict(X_test)
```

```python
# getting the r2_score
r2_score(y_test,y_predict)
```

### Random forest regressor

```python
# creating a random forest model with random forest regressor method
rf_model=RandomForestRegressor(n_estimators=100,max_depth=40,random_state=7)
# fitting the training data
rf_model.fit(X_train,y_train)
```

```python
# storing the predicted value in y_predict
y_predict=rf_model.predict(X_test)
```

```python
# getting the r2_score
r2_score(y_test,y_predict)
```

### so,out of the above regression algorithms, random forest gave the best accuracy of 89.74%

```python
feature_importances = rf_model.feature_importances_
feature_importances
```

### thus, by feature importance, we realise that the two most important features are cement-water ratio,age
