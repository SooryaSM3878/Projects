## Assigning values to X and y

```python
# Defining predictors and target values
X=project_data.iloc[:,2:28] # the features are stored in X
y=project_data.PerformanceRating # the target is stored in y
```

## Label Encoding

```python
 # importing LabelEncoder to convert categorical values to numerical values
from sklearn.preprocessing import LabelEncoder
lab_enc=LabelEncoder() # creating an object named lab_enc for LabelEncoder
```

```python
# using lab_enc we traansform all categorical values into continuous values
X.Gender=lab_enc.fit_transform(X.Gender)
X.EducationBackground=lab_enc.fit_transform(X.EducationBackground)
X.MaritalStatus=lab_enc.fit_transform(X.MaritalStatus)
X.EmpDepartment=lab_enc.fit_transform(X.EmpDepartment)
X.EmpJobRole=lab_enc.fit_transform(X.EmpJobRole)
X.BusinessTravelFrequency=lab_enc.fit_transform(X.BusinessTravelFrequency)
X.OverTime=lab_enc.fit_transform(X.OverTime)
X.Attrition=lab_enc.fit_transform(X.Attrition)
```

## Building a machine learning model to predict employee performance

```python
# importing RandomForestClassifier
from sklearn.ensemble import RandomForestClassifier
# Passing parameters to RandomForestClassifier method and assigning it to pro_model
pro_model=RandomForestClassifier(n_estimators=67,max_depth=32,criterion='gini')
```

```python
# Creating a new variable X1 which contains the important features selected according to correlation and domain knowledge
X1=X.loc[:,['EmpDepartment','EmpJobRole','BusinessTravelFrequency','EmpEnvironmentSatisfaction','YearsSinceLastPromotion','ExperienceYearsAtThisCompany','EmpHourlyRate','EmpJobInvolvement','EmpJobSatisfaction','EmpLastSalaryHikePercent']]
```

```python
# dividing the data into train and test data
from sklearn.model_selection import train_test_split
X1_train,X1_test,y_train,y_test=train_test_split(X1,y,test_size=0.13,random_state=11)
# fitting the training data to the model
pro_model.fit(X1_train,y_train)
```

```python
# giving the test data and storing the predicted value to a variable named y_predict
y_predict=pro_model.predict(X1_test)
```

```python
# to compare the actual result and predicted result, we use accuracy score
# importing accuracy score
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_predict)
```

## Finding the important features

```python
# for getting the important features from the model, we import SelectFromModel from feature_selection subpackage
from sklearn.feature_selection import SelectFromModel
```

```python
# To print the name and gini importance of each feature
for feature in zip(X1,pro_model.feature_importances_):
    print(feature)
```

```python
# The features having greater gini importance values are important features
# Therefore, from the above, we can conclude that the most three important features are: EmpSalaryHikePercent, EmpEnvironmentSatisfaction, YearsSinceLastPromotion
```
