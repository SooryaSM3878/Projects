# a. Do charges of people who smoke differ significantly from the people who don't?

```python
# Smoker vs charges
sb.boxplot(x='smoker',y='charges',data=data)
```

```python
# T-test to test whether charges of smokers differ significantly from those of non-smokers
# Here, null hypothesis H0: charges of smokers and non-smokers are same
# Alternate hypothesis Ha: charges of smokers and non-smokers are different
smoker_charges= data[data['smoker'] == 'yes']['charges']
nonsmoker_charges= data[data['smoker'] == 'no']['charges']
stats.ttest_ind(smoker_charges, nonsmoker_charges)   
```

The above results gives a p-value which is much less than 5%, so reject null hypothesis, and accept alternate hypothesis. Therefore, charges of smokers as well as non-smokers are  significantly different

# b. Does bmi of males differ significantly from that of females?

```python
# Finding the mean bmi of male and female
q2=data.groupby('sex')['bmi']
q2.mean()
```

```python
# Sex vs bmi
sb.boxplot(x='sex',y='bmi',data=data)
```

```python
# T-test to test whether the bmi of males and females are different
# Here, null hypothesis H0: bmi of males and females are same
# Alternate hypothesis Ha: bmi of males and females are different
female_bmi= data[data['sex'] == 'female']['bmi']
male_bmi= data[data['sex'] == 'male']['bmi']
stats.ttest_ind(female_bmi, male_bmi)   
```

Here the p-value>0.05 , i.e; p-value > 5%. So accept H0, i.e; the bmi of males does not significantly differ from that of females

# c. Is the proportion of smokers significantly different in different genders?

```python
# Using chi-square test to find whether the proportion of smokers significanty differ in different genders
# Creating a crosstab to display the distribution of the columns 'sex' and 'smoker' 
crosstab_smokergender=pd.crosstab(data.sex,data.smoker)
df2=crosstab_smokergender.loc[:,['yes']]
df2
```

```python
# Plotting df2
df2.plot(kind='barh',stacked=False,grid=True)
```

```python
# Using chi-square test to find whether the proportion of smokers significanty differ in different genders
# H0 : The proportion of smokers among different genders are same
# Ha: The proportion of smokers among different genders are not same
chi2, p, dof, expected = stats.chi2_contingency(df2)
print("Chi     :",chi2)
print("P-value :",p)
print("dof     :",dof)
print("expected:",expected)
print('\n')
```

Above result shows that p>0.05, which means are the proportion of smokers among different genders are not statistically significant.

# d. Is the distribution of bmi across women with no children, one child and two children, the same ?

```python
# Sorting Values by Sex
data2=data.sort_values(['sex'],ascending=True)
data2
```

```python
filter1 = data2["sex"]=="female"
filter2=data2['children']<3
```

```python
data3=data.where(filter1 & filter2,inplace=False)
```

```python
data3
```

```python
# Droping NaN values
data3.dropna(inplace=True)
```

```python
data3
```

```python
data3.age=data2.age.astype(int)
data3.children=data2.children.astype(int)
```

```python
data3
```

```python
data4=data3.sort_values(['children'],ascending=True)
data4
```

```python
var=['sex','children','bmi']
data5=pd.DataFrame(data4[var])
data5
```

```python
# Using anova test for checking whether the the distribution of bmi across women with no children,1 child and 2children,the same
# Null hypothesis, H0: The distribution of bmi across women with no children, 1 child and 2 children are the same
# Alternate hypothesis, Ha: The distribution of bmi across women with no children, 1 child and 2 children are not same
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
# Entering the model as a regression formula
formula = 'bmi ~ C(children) + C(sex)'
model = ols(formula,data=data5).fit()
model.summary()
```

- Here P=0.676 , which is greater than 0.05
- Therefore, we have to accept the null hypothesis
- Thus, the distribution of bmi across women with no children, 1 child and 2 children are the same
